```python
# there is still quite some redundance, but fixing it is not a priority I assume

def prepare_df_from_txt_num_pts_new( filename_input_num_pts ):
    """
    :param filename_input_num_pts: .txt file with the same structure as
                                   baseline_plots_num_pts_new_ASY.txt
    
    USAGE
    >>> df_num_pts_bl, k_bl, dim_bl, func_list_bl = prepare_df_from_txt_num_pts_new( filename_num_pts_bl )
    
    TODO:
    1. concatenate different measurements with same parameters
    """
    df = pd.DataFrame(columns = {"function", 'k', 'dim', "num_pts", "observations"})

    with open(filename_input_num_pts, 'r') as f:
        for line in f.readlines():
            if( str.isalpha(line[0]) and line[0] != '' ):
                func_name = line.replace("\n", "")
                
            elif len(line.split(" ")) == 2 and line[0] != ' ':
                tmp = line.replace("\n", '' ).strip()
                if tmp != '':
                    k = int(tmp.split(" ")[0])
                    dim = int(tmp.split(" ")[1])
                    
            elif len(line.split(" ")) > 2:
                tmp = line.split(", ")[:-1]
                #print(tmp)
                tmp_obs = list(map(lambda x: float(x), tmp[1:]))
                obs = {"function": func_name, 'num_pts': int(tmp[0]), "k": k, 'dim' : dim,
                       "observations": tmp_obs}
                df = df.append(obs, ignore_index=True)
            else: 
                continue

    # not sure about this one
    df['median'] = df["observations"].apply(lambda x: np.median(x) )
    df['num_rep'] = df["observations"].apply(lambda x: len(x) )
    df['var'] = df["observations"].apply(lambda x: np.var(x) )
    
    df.sort_values(by=['function', "num_pts"], inplace=True)

    assert len(np.unique(df['k'])) == 1, "measurements with multiple values of k" 
    assert len(np.unique(df['dim'])) == 1, "measurements with multiple values of dim" 
    
    if( len(df.drop_duplicates(['function', 'num_pts'])) != len(df) ):
        printf("WARNING: repeated measurements with same parameter values - adapt")
    
    k = np.unique(df['k'])[0]
    dim = np.unique(df['dim'])[0]
    func_list = np.unique(df["function"])
    
    return df, k, dim, func_list


def prepare_df_from_txt_k_new( filename_input_k ):
    """
    :param filename_input_k: .txt file with the same structure as 
                             baseline_plots_num_pts_new_ASY.txt 
                             
    USAGE:
    >>> df_num_pts_bl, num_pts, func_list = prepare_df_from_txt_k_new( filename_k_bl )
    """
    df = pd.DataFrame(columns = {"function", 'k', "num_pts", "observations"})

    with open(filename_input_k , 'r') as f:
        for line in f.readlines():
            if( str.isalpha(line[0]) and line[0] != '' ):
                func_name = line.replace("\n", "")
                
            elif len(line.split(" ")) == 1 and line[0] != ' ':
                tmp = line.replace("\n", '' ).strip()
                if tmp != '':
                    num_pts = int(tmp.split(" ")[0])
                    
            elif len(line.split(" ")) > 1:
                tmp = line.split(", ")[:-1]
                #print(tmp)
                tmp_obs = list(map(lambda x: float(x), tmp[1:]))
                obs = {"function": func_name, 'k': int(tmp[0]), "num_pts": num_pts,
                       "observations": tmp_obs}
                df = df.append(obs, ignore_index=True)
            else: 
                continue


    df['median'] = df["observations"].apply(lambda x: np.median(x) )
    df['var'] = df["observations"].apply(lambda x: np.var(x) )

    func_list = np.unique(df["function"])
    num_pts = np.unique(df['num_pts'])

    return df, num_pts, func_list


def prepare_df_from_txt_dim( filename_input_dim ):
    """
    :param filename_input_dim: .txt file with the same structure as 
                               bl_plots_dim_new_ASY.txt  
                               
    Comment: unlike for prepare_df_from_txt_k_new since
             the pipelines contain the function k-neighbborhood
             both k and dim will affect the runtime 
             => indicate both of them on the plots
             
    USAGE
    >>> df_dim_bl, num_pts_dim_bl,  k_dim_bl, func_list_bl =  prepare_df_from_txt_dim( filename_dim_bl )
    """
    df = pd.DataFrame(columns = {"function", 'dim', "num_pts", "k", "observations"})

    with open(filename_input_dim, 'r') as f:
        for line in f.readlines():
            if( str.isalpha(line[0]) and line[0] != '' ):
                func_name = line.replace("\n", "")

            elif len(line.split(" ")) == 2 and line[0] != ' ':
                tmp = line.replace("\n", '' ).strip()
                if tmp != '':
                    num_pts = int(tmp.split(" ")[0])
                    k = int(tmp.split(" ")[1])

            elif len(line.split(" ")) > 2:
                tmp = line.split(", ")[:-1]
                #print(tmp)
                tmp_obs = list(map(lambda x: float(x), tmp[1:]))
                obs = {"function": func_name, 'num_pts': num_pts, "k": k, 'dim' : int(tmp[0]),
                       "observations": tmp_obs}
                df = df.append(obs, ignore_index=True)
            else: 
                continue

    df['median'] = df["observations"].apply(lambda x: np.median(x) )
    df['var'] = df["observations"].apply(lambda x: np.var(x) )

    assert len(np.unique(df['k'])) == 1, "measurements with multiple values of k" 
    assert len(np.unique(df['num_pts'])) == 1, "measurements with multiple values of num_pts" 
    
    func_list = np.unique(df["function"])
    num_pts = np.unique(df['num_pts'])[0]
    k = np.unique(df['k'])[0]

    return df, num_pts, k, func_list
    
    

def prepare_df_from_txt_k_old( filename_input_k ):
    """
    just copy pasted from one of the old cells
    not sure we need it anymore, but can delete later
    """
    df = pd.DataFrame(columns = {"function", 'k', "num_pts", "observations"})

    with open(filename_input_k, 'r') as f:
        for line in f.readlines():
            if( str.isalpha(line[0]) and line[0] != '' ):
                func_name = line.replace("\n", "")
            elif len(line.split(" ")) == 1 and line[0] != ' ':
                tmp = line.replace("\n", '' ).strip()
                if tmp != '':
                    num_pts = int(tmp)
                    # print(num_pts)
            else:
                tmp = line.split(", ")[:-1]
                tmp_obs = list(map(lambda x: float(x), tmp[1:]))
                obs = {"function": func_name, 'k':int(tmp[0]), "num_pts": num_pts, 
                       "observations": tmp_obs}
                df = df.append(obs, ignore_index=True)

    df.drop_duplicates(['function', 'k'], inplace=True)

    df['median'] = df["observations"].apply(lambda x: np.median(x) )
    df['var'] = df["observations"].apply(lambda x: np.var(x) )

    func_list = np.unique(df["function"])
    num_pts = np.unique(df['num_pts'])[0]

    return df, num_pts
```

## Our First Rooflines

Some of these numbers are just approximations, since it currently we do not account for:

1. exact latency / throughput for certain operations
2. what ports are assigned to what operations 
3. exact implementation of library functions such as `qsort`

I get the info on the latencies / throughput from Aigner's instructions table

**Define performance parameters**

```python
# first number: latency
# second number: 1 / (reciprocal throughput)
# Intel Core i7: Nehalem

fadd = (3, 1)
fcom = (1, 1)
fmul = (5, 1)
fdiv = (14, 1.0/14)     #  7-27, 7-27
fsqrt = (27, 1)
```

**Define general system parameters**

```python
peak_perf = 5 
bandwidth = 64
dim = 4
k = 10
x = np.linspace(0, 50)
beta_line = bandwidth*x

x_intersect1 = float(peak_perf) / bandwidth
x_intersect2 = 4*float(peak_perf) / bandwidth

x_before = np.linspace(0, x_intersect1)
beta_line_before = bandwidth*x_before
x_between = np.linspace(x_intersect1, x_intersect2)
beta_line_between = bandwidth*x_between

x_after = np.linspace(x_intersect2, 50)
```

```python
F3 = (1./10, fcom[1]/fcom[0])
F4 = (1./8, fcom[1]/fcom[0])
#F6 = (1, (k+2)/( k*fadd[0]/fadd[1] + 2*fdiv[0] ) )// same

dim_sample = [ 4, 10, 20, 50, 100 ]

F1_x = [float(dim)/4 for dim in dim_sample]
F1_y = [(4*dim + 1)/(fsqrt[0] + dim*( 3*fadd[0]/fadd[1] + fmul[0]/fmul[1])) for dim in dim_sample]
F1_y_vec = [ 4*f1 for f1 in F1_y ]

k_sample = [10, 15, 20, 40, 100]
F5_x = [ 1 for k in k_sample ]
F5_y = [ (k+2)/(k*fadd[0]/fadd[1] + 2*fdiv[0]) for k in k_sample ]
F5_y_vec = [ 4*f5 for f5 in F5_y ]
```

```python
plt.style.use('seaborn-white') 
from matplotlib import rc
matplotlib.rcParams.update({'font.size': 22})

markersize_plot = 16


#rc('font',**{'family':'serif','serif':['Times'], 'size':22})

f, ax = plt.subplots(figsize=(13, 13))
plt.text(0.01, 5.3, "$\pi=$ {} cycles".format(peak_perf), fontsize=30)
plt.text(0.01, 20, "$4\pi=$ 20 cycles", fontsize=30)
#plt.text(0.45, 21, "P(n) = beta * I(n)", fontsize=30)

title = "Roofline plots for baseline implementation\n\n[flops/cycle]"
ax.set_title(title, loc = "left", fontdict={'fontsize': 30, 'fontweight': 'bold'})

p1 = ax.axhline(y=4*peak_perf, color='b', linestyle='-', linewidth=2)
p2 = ax.axhline(y=peak_perf, color='r',  linestyle='-', linewidth=2)
p3 = ax.plot(x, beta_line, 'black')

plt.vlines(x_intersect1, 0, peak_perf, linestyle='--', color="red")
plt.vlines(x_intersect2, 0, 4*peak_perf, linestyle='--', color="blue")
ax.text(0.01, 0.8, 'Read/write bandwidth', rotation = 55, fontsize=26)

# ORIGINAL
ax.plot(F1_x, F1_y, 'g^', linestyle = "--", markersize = markersize_plot, label = 'PairDist')
ax.plot(F3[0], F3[1], 'g*', markersize = markersize_plot, label = 'K-Dist')
ax.plot(F4[0], F4[1], 'gv', markersize = markersize_plot, label = 'K-Nhood')
ax.plot(F5_x, F5_y, 'go', linestyle = "--", markersize = markersize_plot, label = 'LRD & LOF')

# VECTORIZED

ax.plot(F1_x, F1_y_vec, 'y^', linestyle = "--", markersize = markersize_plot, label = 'PairDist (vec)')
ax.plot(F3[0], 4*F3[1], 'y*', markersize = markersize_plot, label = 'K-Dist (vec)')
ax.plot(F4[0], 4*F4[1], 'yv', markersize = markersize_plot, label = 'K-Nhood (vec)')
ax.plot(F5_x, F5_y_vec, 'yo', linestyle = "--", markersize = markersize_plot, label = 'LRD & LOF (vec)')

# Mark different areas in the roofline plot

ax.fill_between(x_before, beta_line_before, color='C0', alpha=0.2, interpolate=False)
ax.fill_between(x_between, peak_perf, beta_line_between, color='C0', alpha=0.2, interpolate=False)



ax.fill_between(x_after, peak_perf, 4*peak_perf, color='C3', alpha=0.2,
                interpolate=True)
ax.fill_between(x_between, 0, peak_perf, color='C3', alpha=0.2,
                interpolate=True)
ax.fill_between(x_after, 0, peak_perf, color='C3', alpha=0.2,
                interpolate=True)

ax.text(0.0075, 0.1, 'Memory Bound', fontsize=26)
ax.text(1.5, peak_perf + 0.5, 'Compute Bound', fontsize=26)



ax.set_xlim((0.007, 50))
ax.set_ylim((0.07, 30))
ax.set_xlabel(r'$I(n)$ [flops/byte]')
#ax.set_ylabel(r'$P(n)$ [flops/cycle]')
ax.set( xscale="log", yscale="log")

ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=20)

plt.show()
```
